```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Импорт датасета

```python
df = pd.read_csv("/content/drive/MyDrive/Files For Project/AtomPairsData.csv")
df = df.drop(["Unnamed: 0"], axis = 1)
df
```

## Анализ

```python
fig, axs = plt.subplots(1, 2, figsize=(14, 6))  # Создаем 2 графика в ряд

# Первый график
axs[0].scatter(df['304'], df['Ka_mean'])  # Точечный график
axs[0].set_title('Зависимость Ka_mean от свойства фингерпринта 304', fontsize=16)  # Заголовок
axs[0].set_xlabel('304', fontsize=14)  # Подпись оси X
axs[0].set_ylabel('Ka_mean', fontsize=14)  # Подпись оси Y
axs[0].grid(False)  # Удаляем сетку
axs[0].set_xticks([0, 1])  # Ограничиваем значения по оси Y

# Второй график
axs[1].scatter(df['209'], df['Ka_mean'])  # Точечный график
axs[1].set_title('Зависимость Ka_mean от свойства фингерпринта 209', fontsize=16)  # Заголовок
axs[1].set_xlabel('209', fontsize=14)  # Подпись оси X
axs[1].set_ylabel('Ka_mean', fontsize=14)  # Подпись оси Y
axs[1].grid(False)  # Удаляем сетку
axs[1].set_xticks([0, 1])  # Ограничиваем значения по оси Y

plt.tight_layout()  # Подгоняем графики для лучшего отображения
plt.show()
```

```python
print(f'Уникальные значения для 304:', df["304"].value_counts())
print(f'Уникальные значения для 209:', df["209"].value_counts())
```
